```python
from geopandas import GeoSeries, GeoDataFrame, read_file
from shapely.geometry import Point
```

```python
boros = read_file('/home/arshad/Downloads/London_census/london_wards_2011_wgs84.shp')
```

```python
boros
```

```python
boros.set_index('CODE', inplace=True)
boros.sort(inplace=True)
boros
```

```python
boros.reset_index(inplace=True)
boros.set_index('NAME', inplace=True)
boros
```

```python
plt.figure(figsize=(15, 15))
boros.plot()
```

```python
plt.figure(figsize=(15, 15))
boros.plot(alpha=0.0)
boros.geometry.convex_hull.plot()

```

```python
plt.figure(figsize=(60,30))
boros.geometry.area.plot(kind='bar')
plt.tight_layout()
```
