# Selected Indices from 1980-2010

```python
ind = ['NY.GDP.MKTP.KN','EG.IMP.CONS.ZS','SI.DST.10TH.10','TX.VAL.TECH.CD']
ind_name = ['gdp','energy_import','income_10pc','high_tech_exp']
```

```python
dat = wb.download(indicator=ind, country='all', start=1980, end=2010).fillna(0)
dat.columns = ind_name
dat.tail(10)
```

```python
dat.head(10)
```

# Exploratory Analysis

**What trends can we see in the variables?**

```python
dat_01 = dat.reset_index()
```

```python
country_name = dat_01['country'].unique()
country_name = country_name[34:]
```

```python
todrop = dat_01['country'].isin(country_name)
dat_01 = dat_01[todrop]
dat_01.head()
```

```python
dat_01.plot(subplots=True,figsize=(15,25), use_index=False, rot=90, colormap=plt.cm.viridis, kind='line',
           linewidth=6, fontsize=15, x='country', sharex=False)
plt.tight_layout()
```

**What are the trends in the mean of the indicators?**

```python
grouped_mean = dat_01.groupby('country').mean().fillna(0).reset_index()
grouped_mean.head()
```

**The GDP requires normalisation. We will use log-normalisation.**

## Income to GDP Ratio 

```python
inc_gdp = grouped_mean['income_10pc']/grouped_mean['gdp']
sns.distplot(inc_gdp.fillna(0))
```

```python
tech_energy =(grouped_mean['high_tech_exp']/grouped_mean['energy_import']).replace(to_replace=[np.inf,np.nan,-np.inf,
                                                                                              np.NaN, np.NAN,
                                                                                              np.NINF],value=0 )
sns.distplot(tech_energy, kde=True)
```

```python
energy_gdp =(grouped_mean['energy_import']/grouped_mean['gdp']).replace(to_replace=[np.inf,np.nan,-np.inf,
                                                                                              np.NaN, np.NAN,
                                                                                              np.NINF],value=0 )
sns.distplot(energy_gdp, kde=True)
```

```python
grouped_mean.plot(subplots=True,figsize=(15,25), use_index=True,rot=90, colormap=plt.cm.viridis, kind='line',
           linewidth=6, fontsize=15, x='country', sharex=False)
plt.tight_layout()
```

**How are the indicators related?**

```python
plt.figure(figsize=(15,20))
sns.pairplot(grouped_mean, diag_kind="kde")
plt.tight_layout()
```
