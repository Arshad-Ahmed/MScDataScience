# Signal Processing on Graphs

## Import Libraries

```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid", color_codes=True)
import random
random.seed(1111111111111)
plt.rc('axes', grid=False, )
plt.rc('lines', linewidth=4)
```

## Generate some random graph data

```python
G = nx.geographical_threshold_graph(500,10)
```

```python
plt.figure(figsize=(15,15))
nx.draw_spring(G, node_color='#438D80', linewidths=0.01, node_size=100, cmap =plt.cm.magma)
```

## Calculate Degree Centrality

```python
deg_cent = nx.degree_centrality(G)
```

```python
sns.distplot(list(deg_cent.values()))
plt.title("Distribution of Degree Centrality", fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.tight_layout()
```

## Calculate Fourier, Hilbert Transforms etc 

```python
from scipy.fftpack import *
```

```python
deg_cent_val = list(deg_cent.values())
fft_deg = fft(deg_cent_val)
hilbert_deg = hilbert(deg_cent_val)
fft2_deg = fft(fft_deg)
hilbert2_deg = hilbert(hilbert_deg)
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Plot of Degree Centrality and with different transforms", fontsize=25)

plt.subplot(511)
plt.plot(deg_cent_val, label='Degree Centrality' )
plt.legend(loc=1)

plt.ylim(0, max(deg_cent_val)+0.5)
plt.subplot(512)
plt.plot(fft_deg, label='FFT Degree Centrality')
plt.legend(loc=1)

plt.subplot(513)
plt.plot(hilbert_deg, label='Hilbert Degree Centrality')
plt.legend(loc=4)

plt.subplot(514)
plt.plot(fft2_deg, label='Double FFT Degree Centrality')
plt.legend(loc=1)

plt.subplot(515)
plt.plot(hilbert2_deg, label='Double Hilbert Degree Centrality')
plt.legend(loc=1)
```

```python
plt.figure(figsize=(15,6))
plt.scatter(fft_deg,deg_cent_val, label='FFT Deg Cent vs Deg Cent', s=80)
plt.title("Plot of Degree Centrality and Fourier Transform of Degree Centrality", fontsize=25)
plt.xlabel("FFT Values", fontsize=25)
plt.ylabel("Degree Centrality Values", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(loc=1, fontsize=15)
```

```python
plt.figure(figsize=(15,6))
plt.scatter(hilbert2_deg,deg_cent_val, label='Double Hilbert Deg Cent vs Deg Cent', s=60)
plt.title("Plot of Degree Centrality and Double Hilbert Transform of Degree Centrality", fontsize=25)
plt.xlabel("Double Hilbert Transform Values", fontsize=25)
plt.ylabel("Degree Centrality Values", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.legend(loc=1)
plt.legend(loc=1, fontsize=15)
```

```python
from pandas.tools.plotting import autocorrelation_plot
```

Autocorrelation plots are often used for checking randomness in time series. This is done by computing autocorrelations for data values at varying time lags. If time series is random, such autocorrelations should be near zero for any and all time-lag separations. If time series is non-random then one or more of the autocorrelations will be significantly non-zero. The horizontal lines displayed in the plot correspond to 95% and 99% confidence bands. The dashed line is 99% confidence band.

```python
plt.figure(figsize=(15,6))
autocorrelation_plot(fft_deg)
plt.title("Autocorrelation Plot of FFT Degree Centrality", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
```

```python
plt.figure(figsize=(15,6))
autocorrelation_plot(hilbert_deg)
plt.title("Autocorrelation Plot of HT of Degree Centrality", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
```

```python
plt.figure(figsize=(15,6))
autocorrelation_plot(hilbert2_deg)
plt.title("Autocorrelation Plot of Double HT of Degree Centrality", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
```

```python
plt.figure(figsize=(15,6))
autocorrelation_plot(fft2_deg)
plt.title("Autocorrelation Plot of Double FFT Degree Centrality", fontsize=25)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
```
